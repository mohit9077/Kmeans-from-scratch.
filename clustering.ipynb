```python
import numpy as np
from matplotlib import pyplot as plt
from sklearn.datasets import make_blobs
```

```python
X,y=make_blobs(n_samples=500,n_features=2,centers=5,random_state=3)
```

```python
plt.figure(0)
plt.scatter(X[:,0],X[:,1])
plt.grid(True)
plt.show()
## it is a unsupervised lean\rning alogrithm we will not use "y" values anywhere in the code.
```

```python
k=5 ### how many clustres you want to have
  
color=["green","blue","white","yellow","orange"]  ## color for the cluster 
```

```python
## now we have to give random centers for each of the clusters (initialising the random centres)

clusters={}
for i in range(k):
    center= 10*(2*np.random.random((X.shape[1],))-1) ### giving the random centre for each of the cluster
    points=[]
    ## now very center have the points color related to it
    cluster={
        'center':center,
        'points':points, ## for apeending the points which belong to the particular cluster
        'color':color[i] #baiscally the label
        }
    
    clusters[i]=cluster   ## dictionary of dictionary with values point center and color
    
    
    
    
```

```python
clusters
```

```python
## helper function
def distance(x1,x2): 
    return np.sqrt(np.sum((x1-x2)**2))
```

```python
#E-step 
def assignpointtocluster(clusters): 
    for ix in range(X.shape[0]):# for every point in the dataset
        dist=[]
        curr_x=X[ix]
        #print(curr_x)
        
        
        for kx in range(k):
            d=distance(curr_x,clusters[kx]['center'])
            #print(d)
            dist.append(d)

            ## so here we are calculating distance of ecah point with each cluster center 
        current_cluster=np.argmin(dist) 
        #print(current_cluster)
        clusters[current_cluster]['points'].append(curr_x)
        ##add the point to the cluster which has minimum distance from list of 5 clusters

    
    ### this is also known as E-step
    ## expectation maximazie our algorithm
    
    
 #CLEAR THE LIST AS CENTERS ARE NOW HAVE BEEEN CHANGED 
## here we update every cluster center aaccording to the mean
def update_clusters(clusters): 
    for kx in range(k):
        pts=np.array(clusters[kx]['points'])
        
        if(pts.shape[0])>0:
            new_u=pts.mean(axis=0)
            clusters[kx]['center']=new_u  ## updating the clusters with new_cluster_centre 
            clusters[kx]['points']=[]     ## empty the lists of the points as new points will be added in the E-step for 
                                          ## cluster center 
        
                
                
```

```python
def plot_clusters(clusters):
    
    for kx in range(k):
        print(len(clusters[kx]['points']))
        pts=np.array(clusters[kx]['points'])
        #print(pts.shape)
        
        ## here it is possible that any of the cluster centre may no receive any point
        ## so to solve this
        
        try:
            plt.scatter(pts[:,0],pts[:,1],color=clusters[kx]['color'])
        except:
            pass
        
        
        # plotting the center
        uk=clusters[kx]['center']
        plt.scatter(uk[0],uk[1],color='black',marker='*')
```

```python
assignpointtocluster(clusters)
```

```python
plot_clusters(clusters)
```

```python
update_clusters(clusters)
```

```python
assignpointtocluster(clusters)
plot_clusters(clusters)
update_clusters(clusters)
```
